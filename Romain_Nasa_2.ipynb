Absolute Peaks: Identify and visualize periods where the absolute rate of change for "kp" or "norm" is highest, indicating significant movement.

Aligned Movement: Identify periods where "kp" and "norm" are both moving significantly in the same direction (either both increasing or both decreasing).

Threshold-Based: Identify periods where the rate of change exceeds a certain threshold, indicating significant movement.
