```python
kp_val = "/content/drive/My Drive/NASA Challenge/KP_Values_Dina.xlsx"
kp17 = pd.read_excel(kp_val, '2017', parse_dates = [11])
kp17
```

```python
kp17.info()
```

```python
kp17['Timestamp'] = kp17['Date'] + pd.to_timedelta(kp17['hour'], unit='h') + pd.to_timedelta(kp17['min'], unit='m')
kp17['Timestamp'] = kp17['Timestamp'].dt.floor('Min')
kp17
```

```python
col_kp17 = ['Timestamp', 'Actual Kp', 'Ap', 'Range N']
kp17_clean = kp17[col_kp17]
kp17_clean
```

```python
kp17_clean.to_excel('kp_clean_2017.xlsx', index=False)
```
